```python
category_dict_filepath = os.path.join( './category_dict.json')


with open(category_dict_filepath, 'r') as f:
        category_dict = json.load(f)

def clean_alias(string):
    return string.lower().strip()

def define_categories(categories):
    new_categories = set()
    for category in categories:
        new_categories.add(lookup_category(category))
        
    return list(sorted(new_categories))

def lookup_category(category):
    category = clean_alias(category)
    return category_dict[category]

def lookup_category_id(category):
    category = clean_alias(category)
    return category_dict_id[category]
```

```python
tags = set()
for x in data:
    for tag in x.get('tags', []):
        tag.lower().strip()
        tags.add(tag)
#print(tags)
# Look up category mappings and return the used categories
defined_categories = define_categories(categories=tags)
# Map tags to cagegories and define "categories" key
for x in data:
    x['categories'] = set()
    for tag in x.get('tags', []):
        x['categories'].add(lookup_category(category=tag))
    x['categories'] = list(x['categories'])
    
    

# Create category to idx mapping dicts
category2idx = {x: i for i, x in enumerate(defined_categories)}
idx2category = {i: x for i, x in enumerate(defined_categories)}
```
